```python
# create function based on @odi swissparlpy package for downloading batches of data 
def save_transcripts_of_session(id, path):
    if not os.path.exists(path):
        os.mkdir(path)
    data = sp.get_data("Transcript", LanguageOfText="DE", Language="DE", CantonId=id)
    print(f"{data.count} rows loaded.")
    df = pd.DataFrame(data)
    pickle_path = os.path.join(path, f'{id}.pks')
    df.to_pickle(pickle_path)
    print(f"Saved pickle at {pickle_path}")
```
